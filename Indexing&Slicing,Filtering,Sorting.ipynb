# Filtering

```python
dataset = pd.read_csv("/content/sample_data/california_housing_test.csv")
dataset
```

```python
dataset["housing_median_age"] <= 20
```

```python
housing_20 = dataset[dataset["housing_median_age"] <= 20]
```

```python
housing_20["housing_median_age"].min()
```

```python
housing_20["housing_median_age"].max()
```

```python
%%time
a = dataset[(dataset["housing_median_age"] <= 20) &
        (dataset["population"] < 100)]
```

# query

```python
%%time
b = dataset.query("housing_median_age <= 20 and population < 100 ")
```

```python
a == b
```

# Sorting

```python
# Mengurutkan data "housing-median_age" dari kecil ke besaar (True)
dataset.sort_values(by = "housing_median_age", ascending = True).head(5)
```

```python
# Mengurutkan dari data "housing-median_age" dari kecil ke besaar (True) dana "Poplation" dari besar ke kecil (False)
dataset.sort_values(by = ["housing_median_age", "population"],
                    ascending = [True, False]).head(5)
```
